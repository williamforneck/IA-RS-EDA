## Análise Bivariada

```python
# Hipótese: Clientes com contrato mensal tem maior churn
# Contruir uma tabela de contingencia entre tipo de contrato e churn - quantidade
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='Total')
```

```python
# Hipótese: Clientes com contrato mensal tem maior churn
# Contruir uma tabela de contingencia entre tipo de contrato e churn - proporção
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='Total', normalize='index')
```

Considerando o publico total, os clientes de conntrato mensal representam 55%. Porém, se considerarmos apenas os clientes que abandonaram o serviço, 88% tinham contrato mensal. Desta forma, podemos afirmar que há uma correlação entre o tipo de contrato mensal e o abandono.

```python
# Avaliar a correlação entre duas variaveis categoricas (qualitativas)
# Executar um teste de hipotese chamado de Teste (Chi-Square) Qui-Quadrado de Pearson
# Num teste de hipotese, a hipotese nula é a hipotese que queremos rejeitar
# h0 (hipotese nula) = As variaveis são independentes
# h1 (hipotese complementar) = As variaveis não são independentes

# O Teste serve para confirmar ou rejeitar a hipotese nula
# Quando a probabilidade ao observarmos h0 é menor que 5% (0.05) (p-value) rejeitamos h0
# Recusamos a hipotese nula e aceitamos a hipotese complementar
```

```python
# Gerar um DDF da cross tab (sem totais)
df_cross_tab_churn_contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
```

```python
# Calcular os Scores e P-Value
from scipy.stats import chi2_contingency
```

```python
# Tupla = lista imutavel
chi_scores_churn_contract = chi2_contingency(df_cross_tab_churn_contract)
```

```python
chi_scores_churn_contract
```

```python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
# Apresentar numeros com ddecimais sem a notação cientifica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
# Criar DF com os resultados
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': scores_churn_contract, 'p-Value': pvalues_churn_contract})
```

```python
# Analisar scores e p-values
df_chi_scores_churn_contract
```

Conforme Teste estatístico (Chi-Square), o p-value <= 0.05, desta forma rejeitamoss a hipotese nula, ou seja, as variaveis não são independentes. Pelo Qui2 alto, podemos afirmar que há uma forte correlação.

```python
# Avaliar a correlação entre uma  variavel qualitativa e uma quantitativa
# Usar o mesmo teste de Chi-Square, mas transformar a variavel quantitativa em qualitativa
# Hipótese: Clientes com menos de 6 meses de contrato tem maior churn

# Criar uma nova coluna com a variavel qualitativa (yes para menos de 6 meses e no para mais de 6 meses)    
# coluna = 'TempoMenor6Meses'

import numpy as np
```

```python
df_churn['TempoMenor6Meses'] = np.where(df_churn.tenure < 6, 'yes', 'no')
```

```python
df_churn.head(5)
```

```python
df_cross_tab_churn_tenure = pd.crosstab(df_churn.Churn, df_churn.TempoMenor6Meses)
```

```python
# Calcular os Scores e P-Value
chi_scores_churn_tenure = chi2_contingency(df_cross_tab_churn_tenure)
```

```python
scores_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
pvalues_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
```

```python
# Criar DF com os resultados
df_chi_scores_churn_tenure = pd.DataFrame({'Qui2': scores_churn_tenure, 'p-Value': pvalues_churn_tenure})
```

```python
# Analisar os resultados
df_chi_scores_churn_tenure
```

Como o P-value <= 0.05, rejeitamoss a H0, ou seja, as variáveis não são independentes. E vale mencionar que com base no Qui2 resultantte das 2 analises, a correlação entre Churn e Tempo dde Contrato < 6 meses é menor do que a correlação entre Churn e Tipo de contrato

## Detecção de Outliers (Valores atípicos)

```python
# Remover linhas com valores nulos
df_churn.dropna(axis=0, inplace=True)
```

```python
# Box Plot geral
df_churn.TotalCharges.plot.box()
```

```python
# Box Plot agrupado por Contract
df_churn.plot.box(column='TotalCharges', by='Contract')
```

```python
# Criar DF somente com clientes Month to month
df_churn_month = df_churn[df_churn.Contract == 'Month-to-month']
```

```python
len(df_churn_month)
```

```python
# Histograma - Checar visualmente se a distribuição é normal
df_churn_month.TotalCharges.plot.hist() 
```

```python
# Método de Tukey - Detectar outliers - IQR (Distruibuições não normais)
# IQR - RANGE INTERQUARTIL
# IQR = Q3 - Q1
# Q1 = 25% dos dados
# Q3 = 75% dos dados
q1_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.25)
q3_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.75)
iqr_TotalCharges_month = q3_TotalCharges_month - q1_TotalCharges_month
iqr_TotalCharges_month
```

```python
# Limite Inferior e Superior
limite_inferior_TotalCharges_month = q1_TotalCharges_month - (iqr_TotalCharges_month * 1.5)
limite_superior_TotalCharges_month = q3_TotalCharges_month + (iqr_TotalCharges_month * 1.5) 
limite_inferior_TotalCharges_month, limite_superior_TotalCharges_month
```

```python
len(df_churn_month[(df_churn_month.TotalCharges < limite_inferior_TotalCharges_month) | (df_churn_month.TotalCharges > limite_superior_TotalCharges_month)])
```

### ZScore

```python
# ZScore (usado para distribuições normais)
# Indica quantos desvios padrões um ponto específico esta da media
# z = (x - media) / desvio padrao
# qualquer valor que seja maior que 3 ou menor que -3 é considerado outlier
from scipy.stats import zscore
```

```python
z = np.abs(zscore(df_churn_month.TotalCharges))
z
```

```python
df_churn_month[z > 3.0]
```
